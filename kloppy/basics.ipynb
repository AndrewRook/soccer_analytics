```python
from mplsoccer.pitch import Pitch

home_team = data.metadata.teams[0]

pitch = Pitch(#pitch_type='statsbomb',
              pitch_color='white', line_color='#c7d5cc')
fig, ax = pitch.draw()

def get_color(event, player):
    if player == event.player:
        return "blue"
    elif player.team == event.player.team:
        return "green"
#     elif player.position.position_id == '1':
#         return "black"
    else:
        return "red"
    

x, y, color = zip(*[
  (coordinates.x, coordinates.y, get_color(throw_in, player))
     for player, coordinates in throw_in.freeze_frame.players_coordinates.items()
])

_ = pitch.scatter(x, y, color=color, ax=ax)
```
